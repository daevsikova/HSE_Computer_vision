```python
from itertools import permutations

def img_permutations(image):
    channels_p = permutations([0, 1, 2])
    image = np.array([image[:, :, p] for p in channels_p])
    image = np.vstack([np.hstack(image[:2]), np.hstack(image[2:4]), np.hstack(image[4:])])
    return image


images = img_permutations(lena)
plt.figure(figsize=(20,10))
plt.imshow(images)
plt.show()
```

##### Задача 10 (3 балла)
Переведите изображение **lena** из RGB в CIE Lab при помощи встроенной функции OpenCV. Загрузите исходное иображение еще раз и переведите из RGB в CIE Lab самостоятельно с помощью формул (google it). Сравните полученные изображения, построив разницу между ними. (Должен получится ~0)

```python
lena_opencv = cv2.cvtColor(lena, cv2.COLOR_RGB2Lab)
plt.imshow(lena_opencv)
plt.show()
```

```python
def gamma(image):
    image = image / 255.
    condition = (image > 0.04045)
    image[condition] = np.power((image[condition] + 0.055) / 1.055, 2.4)
    image[~condition] = image[~condition] / 12.92
    return image


def rgb2lab(image):
    image = gamma(image)
    transf_matrix = np.array([[0.412453, 0.357580,  0.180423],  
                             [0.212671, 0.715160, 0.072169], 
                             [0.019334, 0.119193, 0.950227]])

    xyz = np.dot(image, transf_matrix.T)
    xyz[:,:,0] = xyz[:,:,0] / 0.950456
    xyz[:,:,2] = xyz[:,:,2] / 1.088754

    cond = (xyz[:, :, 1] > 0.008856)
    L = np.zeros_like(xyz[:, :, 1])
    L[cond] = xyz[:, :, 1][cond] ** (1 / 3) * 116 - 16
    L[~cond] = xyz[:, :, 1][~cond] * 903.3

    cond = (xyz > 0.008856)
    xyz[cond] = xyz[cond] ** (1 / 3) 
    xyz[~cond] = xyz[~cond] * 7.787 + 16/116 

    a = 500 * (xyz[:, :, 0] - xyz[:, :, 1])
    b = 200 * (xyz[:, :, 1] - xyz[:, :, 2])


    lab = np.stack([L * 255/100, a + 128, b + 128], axis=-1).astype('int')
    return lab
```

```python
lena_mine = rgb2lab(lena)
plt.imshow(rgb2lab(lena))
plt.show()
```
